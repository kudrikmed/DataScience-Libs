# Задание 1

```python
import numpy as np
```

```python
a = np.array([[1, 6],[2, 8],[3, 11],[3, 10],[1, 7]])
```

```python
mean_a = np.mean(a, axis = 0)
```

```python
mean_a
```

# Задание 2

```python
a_centered = a - mean_a
a_centered
```

# Задание 3

```python
a_centered_sp = sum(np.prod(a_centered, axis=1))
print(a_centered_sp / (len(a_centered) - 1))
```

# Задание 4

```python
a_centered_sp / (len(a_centered) - 1) == np.cov(a.T)[0, 1]
```
